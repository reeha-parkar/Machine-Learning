# **Write a program to implement k-Nearest Neighbor algorithm to classify the weather dataset. Print confusion matrix. (k=5)**

### **Import statements**

```
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split 
from sklearn import preprocessing
```

### **Reading the csv file**

```
data = pd.read_csv('weather.csv')
data.head()
```

### **Checking for null values**

```
data.isnull().sum()
```

### **Shape of dataframe**

```
data.shape
```

### **Dropping all null and missing values**

```
data = data.dropna()
data
```

(328, 22) is the shape after removing all null values which does not affect our dataset much. Earlier, it was (366, 22)

### **Encoding the labels of the output to from (yes, no) to (1, 0)**

```
data.replace('Yes', '1', inplace=True)
data.replace('No', '0', inplace=True)
data
```

### **Encoding the columns with string elements to numerical (int64) values**

```
unique_dir = list(pd.unique(data['WindGustDir']))
print(unique_dir)
```

```
data = data.replace({'WindGustDir': {'NW': 1, 'ENE': 2, 'SSE': 3, 'SE': 4,'E': 5, 'S': 6,'N': 7, 'WNW': 8,'ESE': 9, 'NE': 10,'NNE': 11, 'NNW': 12,'SW': 13, 'W': 14,'WSW': 15, 'SSW': 16}}) 
data
```

### **Getting all numerical values**

```
data = data.replace({'WindDir3pm': {'NW': 1, 'ENE': 2, 'SSE': 3, 'SE': 4,'E': 5, 'S': 6,'N': 7, 'WNW': 8,'ESE': 9, 'NE': 10,'NNE': 11, 'NNW': 12,'SW': 13, 'W': 14,'WSW': 15, 'SSW': 16}}) 
data = data.replace({'WindDir9am': {'NW': 1, 'ENE': 2, 'SSE': 3, 'SE': 4,'E': 5, 'S': 6,'N': 7, 'WNW': 8,'ESE': 9, 'NE': 10,'NNE': 11, 'NNW': 12,'SW': 13, 'W': 14,'WSW': 15, 'SSW': 16}}) 
data
```

### **Checking the data types, if they are all numerical**

```
data.dtypes
```

### **Converting object data types to integer type**

```
data['RainToday'] = data['RainToday'].astype(int)
data['RainTomorrow'] = data['RainTomorrow'].astype(int)
data
```

### **Final check of all data types**

```
data.dtypes
```

### **Defining x (input features) and y (output variable)**

where all the columns are considered for KNN Algorithm and y = RainTomorrow will predict if there will be rainfall tomorrow

```
x = data.drop(['RainTomorrow'], axis=1)
y = data['RainTomorrow']
y.head()
```

### **Implementing the KNNClassifier Model for k=5**

```
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
```

### **Splitting the data**

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
model.fit(x_train, y_train)
```

### **Predicting the output**

```
MinTemp = 8.0
MaxTemp	= 24.3
Rainfall = 0.0
Evaporation	= 3.4
Sunshine = 	6.3
WindGustDir = 1
WindGustSpeed = 30.0	
WindDir9am = 13	
WindDir3pm = 1
WindSpeed9am = 	6.0
WindSpeed3pm = 	20
Humidity9am	= 68
Humidity3pm	= 29
Pressure9am	= 1019.7
Pressure3pm	= 1015.0
Cloud9am = 7
Cloud3pm = 7
Temp9am = 14.4
Temp3pm = 23.6
RainToday = 0	
RISK_MM = 3.6
print(model.predict([[MinTemp, MaxTemp, Rainfall, Evaporation, Sunshine, WindGustDir, WindGustSpeed, WindDir9am, WindDir3pm, WindSpeed9am, WindSpeed3pm, Humidity9am, Humidity3pm, Pressure9am, Pressure3pm, Cloud9am, Cloud3pm, Temp9am, Temp3pm, RainToday, RISK_MM]]))
```

### **Accuracy of the model**

```
y_pred = model.predict(x_test)
score = metrics.accuracy_score(y_test,y_pred)
print(score*100,'%')
```

### **Confusion Matrix**

```
from sklearn.metrics import confusion_matrix

print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```
