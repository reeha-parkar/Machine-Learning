```python
import pydotplus
from sklearn.datasets import load_iris
from sklearn import tree
from IPython.display import Image, display
```

```python
def load_data_set():
    iris = load_iris()
    return iris
```

```python
def train_model(iris):
    clf = tree.DecisionTreeClassifier()
    clf = clf.fit(iris.data, iris.target)
    return clf
```

```python
def display_image(clf,iris):
    dot_data = tree.export_graphviz(clf,out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
    graph = pydotplus.graph_from_dot_data(dot_data)
    display(Image(data=graph.create_png()))
```

```python
if __name__ == '__main__':
    iris_data = load_iris()
    decision_tree_classifier = train_model(iris_data)
    display_image(clf=decision_tree_classifier, iris=iris_data)
```
